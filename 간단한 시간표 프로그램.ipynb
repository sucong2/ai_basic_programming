### 단순한 시간표 관리 프로그램 만들기

```python
# 프로그램 제목 출력
print(' [단순한 수업시간표 관리 프로그램]')
print('===========================')

# 수업정보 초기화
days = 5  # 요일 수
hours = 8 # 시간 수

classes = [[' ' for j in range(days)] for i in range(hours)] 
        
# 수업정보 예시
# classes[4][1] = 'A' : 화(1)요일 5교시(5-1) 과목 'A' 수업 

# 요일 문자열
weekdays = ['월', '화', '수', '목', '금'] 

while True:
    
    # 시간표 표시
    print("\n")
    print("[ TimeTable ]")
    print("-------------")
    print("| |M|T|W|T|F|")                # 요일 출력
    print("-------------")
    
    for i in range(hours):
        print('|', end='')
        print(str(i+1), end='|')          # 교시 출력
        for j in range(days):
            print(classes[i][j], end='|') # 알파벳 출력
        print()
    print("-------------\n")
    
    # 프로그램 메뉴  
    choice = input(">>> 메뉴를 선택하세요 (1. 수업등록, 2. 종료): ")

    if choice == '2':   # 프로그램 종료
        break
    elif choice != '1': # 다른 입력 무시
        print(">>> [ERROR] 잘못 입력하셨습니다!")  # 오류 처리
        continue
        
    # 수업의 요일 및 교시 입력      
    wday, shour = input(">>> 등록할 수업의 요일과 교시를 입력하세요: ").split()
    
    # 요일 입력 오류 체크
    if weekdays.count(wday) == 0:
        print(">>> [ERROR] 요일을 잘못 입력하셨습니다!")      # 요일 입력 오류 
        continue
    
    wday = weekdays.index(wday)   # 요일의 위치 찾기    
    shour = int(shour)-1          # 시작 시간 인덱스
    
    # 수업시간 범위 체크
        
    if not (0<=shour<=7):
        print(">>> [ERROR] 시작 시간을 잘못 입력하셨습니다!")  # 시작 시간 입력 오류     
        continue
    
    # 수업의 연강 시간 입력  
    count = int(input(">>> 등록할 수업의 연강시간을 입력하세요: "))

    # 연강 시간 범위 체크
    if count < 1:
        print(">>> [ERROR] 연강 시간을 잘못 입력하셨습니다!")  # 연강 시간 입력 오류
        continue
    elif not (0<=shour+count-1<=7):
        print(">>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!")  # 범위 오류
        continue

    dup = False   # 중복 체크 변수    
    # 수업 중복 체크
    for i in range(count):
        if classes[shour+i][wday] != ' ':
            print(">>> [ERROR] 해당 시간에 이미 수업이 있습니다!")  # 범위 오류    
            dup = True
            break
            
    if dup:       # 중복 시 처음으로
        continue
        
    # 수업의 알파벳 입력  
    title = input(">>> 등록할 수업의 알파벳을 입력하세요: ")        

    # 수업 정보 입력
    for i in range(count):
        classes[shour+i][wday] = title[0]  # 맨 처음 문자만 입력
```
