####다음과 같이 numpy 테이터를 입력변수로 받는 func1(X), func2(X) 함수는 입력변수 X 모든 원소 값을 출력하는 기능을 가지고 있음. 
####예를 들어 입력변수 타입이 행렬이고 각 원소값이 다음과 같다면 func1, func2 함수는 다음과 같이 출력함.
####(1) 인덱스를 이용하여 입력변수 모든 원소 값을 출력하도록 func1(var) 구현하시오.
####(2) iterator 이용하여 입력변수 모든 원소 값을 출력하도록 func2(var) 구현하시오.

image.png

```python
import numpy as np
# index 이용하여 넘파이 원소값 출력
def func1(var):
	if var.ndim == 1: # vector
		for index in range(len(var)):
			print('value = ', var[index])
	elif var.ndim == 2: # matrix
		rows = var.shape[0]
		columns = var.shape[1]
		for row in range(rows):
			for column in range(columns):
				print('value = ', var[row, column])
	else:
		print('not support tensor...')

# iterator 이용하여 넘파이 원소값 출력
def func2(var):
	it = np.nditer(var, flags=['multi_index'], op_flags=['readwrite'])
	while not it.finished:
		index = it.multi_index
		print('value = ', var[index])
		#print('index = ', index, ', value = ', var[index])
		it.iternext()

X1 = np.array([1, 2, 3, 4, 5]) # vector
X2 = np.array([ [2, 4], [1, 2], [0, 5] ]) # matrix
X3 = np.array([ [ [2, 4], [1, 2] ], [ [3, 5], [6, 8] ] ]) # tensor

# vector X1 에 대한 func1, func2 호출
func1(X1)
print('==============')
func2(X1)

# matrix X2 에 대한 func1, func2 호출
func1(X2)
print('==============')
func2(X2)

# tensor X3 에 대한 func1, func2 호출
func1(X3)
print('==============')
func2(X3)
```

####다음의 조건을 만족하는 함수 loss_func 구현하시오. 여기서 
####xdata = np.array([1, 2, 3, 4, 5]).reshape(5,1) 
####tdata = np.array([3, 4, 5, 6, 7]).reshape(5,1)
####y = xdata ∙ W + b 이며 W.shape = (1,1), b.shape = (1,) 형상을 가지며 임의의 값으로 초기화 되어 있다.

image.png

```python
import numpy as np

W = np.random.rand(1,1)
b = np.random.rand(1)
print(W, b)

def loss_func(xdata, tdata):
	y = np.dot(xdata, W) + b
	loss = np.sum((tdata-y)**2) / (len(xdata))
	return loss

xdata = np.array([1, 2, 3, 4, 5]).reshape(5,1)
tdata = np.array([3, 4, 5, 6, 7]).reshape(5,1)
ret = loss_func(xdata, tdata)
print('ret = ', ret)
```

####matplotlib 이용하여 아래와 같은 sin(x), cos(x) 그래프를 출력하시오.
####여기서 x = np.arange(0, 15, 0.1) , y1 = np.sin(x), y2 = np.cos(x)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0, 15, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.title('sin(x) / cos(x)')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.grid()
plt.xlim(0, 15)
plt.ylim(-1,1)
plt.plot(x, y1, label='sin(x)', color='r', ls='--')
plt.plot(x, y2, label='cos(x)', color='g', ls='--')
plt.legend(loc='best')
plt.show()
```
