```python
def personal_info(name, age):
    print('이름: ', name)
    print('나이: ', age)
```

```python
x = {'name': '홍길동', 'age': 30}
```

```python
personal_info(*x)
```

```python
personal_info(**x)
```

```python
personal_info(**{'name': '홍길동', 'age': 30})
```

```python
def personal_info(**kwargs):
    for kw, arg in kwargs.items():
        print(kw, ': ', arg)
```

```python
personal_info(name='홍길동', age=30)
```

```python
personal_info(name='홍길동', age=30, address='구미시')
```

```python
personal_info('홍길동', 30, '구미시')
```

```python
personal_info(**x)
```

```python
def personal_info(**kwargs):
    if 'name' in kwargs:
        print('name: ', kwargs['name'])
    if 'age' in kwargs:
        print('age: ', kwargs['age'])        
```

```python
personal_info(**x)
```

```python
def personal_info(name, age, address='비공개'):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)
```

```python
personal_info(name='홍길동', age=30)
```

```python
personal_info(name='홍길동', age=30, address='구미시')
```

```python
def personal_info(name, age, address='비공개', gender='남'):
    print('이름: ', name)
    print('나이: ', age)
    print('주소: ', address)
    print('성별: ', gender)    
```

```python
personal_info(name='홍길동', age=30)
```

```python
personal_info(name='홍길동', age=30, gender='여')
```

```python
def hello(count):
    if count == 0:
        return
    
    print('Hello, World!', count)
    count -= 1
    
    hello(count)
```

```python
hello(5)
```

```python
hello(10)
```
